### Sensitivity Analysis on Liquid Flowrate
**The Influence of Changing Liquid Flowrate on CO2 Absorption**

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
# import seaborn as sns
# sns.set()
```

```python
z  = np.linspace(0, 0.5, 1000)
```

**Reading the data** 
- data_1: flowrate = 0.1 ml/min "b"
- data_2: flowrate = 0.3 ml/min "g"
- data_3: flowrate = 0.45 ml/min "y"
- data_4: flowrate = 0.6 ml/min "m"

**Read Files From the Directory**

```python
data_1 = pd.read_csv("flowrate_0.1_.csv")
data_2 = pd.read_csv("flowrate_0.3_.csv")
data_3 = pd.read_csv("flowrate_0.45_.csv")
data_4 = pd.read_csv("flowrate_0.6_.csv")
data_1.head() #just to realize the header
```

**Read Files From GitHub Repository**

```python
data_1 = pd.read_csv("https://raw.githubusercontent.com/jeff-ball/A-Python-Tutorial-with-a-Code/main/flowrate_0.1_.csv")
data_2 = pd.read_csv("https://raw.githubusercontent.com/jeff-ball/A-Python-Tutorial-with-a-Code/main/flowrate_0.3_.csv")
data_3 = pd.read_csv("https://raw.githubusercontent.com/jeff-ball/A-Python-Tutorial-with-a-Code/main/flowrate_0.45_.csv")
data_4 = pd.read_csv("https://raw.githubusercontent.com/jeff-ball/A-Python-Tutorial-with-a-Code/main/flowrate_0.6_.csv")
data_4.head() #just to realize the header
```

```python
plt.figure(figsize=(15, 15))
plt.subplot(2, 2, 1)
plt.xlabel("channel length (m)", fontsize=15)
plt.ylabel("Conversion of CO2%", fontsize=15)
plt.title("Conversion of CO2 vs. Channel Length", fontweight = "bold")
plt.xlim(0,0.5)
plt.ylim(0,100)
plt.plot(z,data_1["CO2 conversion"],'b',label ="0.1 ml/min")
plt.plot(z,data_2["CO2 conversion"],'g',label ="0.3 ml/min")
plt.plot(z,data_3["CO2 conversion"],'y',label ="0.45 ml/min")
plt.plot(z,data_4["CO2 conversion"],'m',label ="0.6 ml/min")
legend = plt.legend(loc='upper left', shadow=True, fontsize='large')


plt.subplot(2, 2, 2)
plt.xlabel("channel length (m)", fontsize=15)
plt.ylabel("Conversion of MEA%", fontsize=15)
plt.title("Conversion of MEA vs. Channel Length", fontweight = "bold")
plt.xlim(0,0.5)
plt.ylim(0,100)
plt.plot(z,data_1["MEA conversion"],'b',label ="0.1 ml/min")
plt.plot(z,data_2["MEA conversion"],'g',label ="0.3 ml/min")
plt.plot(z,data_3["MEA conversion"],'y',label ="0.45 ml/min")
plt.plot(z,data_4["MEA conversion"],'m',label ="0.6 ml/min")
legend = plt.legend(loc='upper left', shadow=True, fontsize='large')


plt.subplot(2, 2, 3)
plt.xlabel("channel length (m)", fontsize=15)
plt.ylabel("Concentration of CO2 gas (mol/m3)", fontsize=15)
plt.title("Concentration of CO2 gas vs. Channel Length", fontweight = "bold")
plt.xlim(0,0.5)
plt.ylim(0,40)
plt.plot(z,data_1["CO2G"],'b',label ="0.1 ml/min")
plt.plot(z,data_2["CO2G"],'g',label ="0.3 ml/min")
plt.plot(z,data_3["CO2G"],'y',label ="0.45 ml/min")
plt.plot(z,data_4["CO2G"],'m',label ="0.6 ml/min")
legend = plt.legend(loc='lower left', shadow=True, fontsize='large')


plt.subplot(2, 2, 4)
plt.xlabel("channel length (m)", fontsize=15)
plt.ylabel("Concentration of MEA (mol/m3)", fontsize=15)
plt.title("Concentration of MEA vs. Channel Length", fontweight = "bold")
plt.xlim(0,0.5)
plt.ylim(0,1000)
plt.plot(z,data_1["MEA"],'b',label ="0.1 ml/min")
plt.plot(z,data_2["MEA"],'g',label ="0.3 ml/min")
plt.plot(z,data_3["MEA"],'y',label ="0.45 ml/min")
plt.plot(z,data_4["MEA"],'m',label ="0.6 ml/min")
legend = plt.legend(loc='lower left', shadow=True, fontsize='large')


plt.tight_layout()
plt.show()
```
